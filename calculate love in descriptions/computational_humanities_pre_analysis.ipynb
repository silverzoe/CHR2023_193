```python
# read 3 data and combine them into one dataframe, only keep information for main analysis
import pandas as pd
fiction = pd.read_csv("fiction.csv", encoding = 'utf-8')
fiction.info()
```

```python
# change year into uniformed year
fiction['Year'] = fiction['Year'].str[:4]
fiction
```

```python
import jieba
import re 

#transfer columns into str type
fiction['Baidu_description'] = fiction['Baidu_description'].astype(str)
fiction['Description_scholar'] = fiction['Description_scholar'].astype(str)
fiction['Wiki_description'] = fiction['Wiki_description'].astype(str)


def cutword(x):
    seg = jieba.cut(x)
    return ' '.join(seg)

fiction['baidu_cutted'] = fiction['Baidu_description'].apply(cutword)
fiction['scholar_cutted'] = fiction['Description_scholar'].apply(cutword)
fiction['wiki_cutted'] = fiction['Wiki_description'].apply(cutword)

fiction
```

```python
# remove punctuation in this step
# split tokenizaed text so that it can be calculated by length
fiction['baidu_cutted'] = fiction['baidu_cutted'].str.split()
fiction['scholar_cutted'] = fiction['scholar_cutted'].str.split()
fiction['wiki_cutted'] = fiction['wiki_cutted'].str.split()


#remove the punctuation chinese by using package zhon.
from zhon.hanzi import punctuation
punctuation_str = punctuation
for i in punctuation:
    fiction['baidu_cutted'] = fiction['baidu_cutted'].astype(str).str.replace(i, '')
    
for i in punctuation:
    fiction['scholar_cutted'] = fiction['scholar_cutted'].astype(str).str.replace(i, '')
for i in punctuation:
    fiction['wiki_cutted'] = fiction['wiki_cutted'].astype(str).str.replace(i, '')

fiction
```

```python
# here we calculate the totol length of each description 
fiction['baidu_length'] = fiction['baidu_cutted'].str.len()
fiction['wiki_length'] = fiction['wiki_cutted'].str.len()
fiction['scholar_length'] = fiction['scholar_cutted'].str.len()
```

```python
fiction[["baidu_length","wiki_length","scholar_length"]] = fiction[["baidu_length","wiki_length","scholar_length"]].replace(7, 0)
```
